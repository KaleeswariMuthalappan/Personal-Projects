```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv',encoding = 'unicode_escape') #to remove the encoding error
```

## Data Cleaning

```python
df.drop(['Status','unnamed1'],axis=1,inplace = True) #Removing the unwanted columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
#Type case to int
df['Amount']=df['Amount'].astype('int')
```

```python
df.info()
```

```python
#Renaming the Column name
df.rename(columns = {"Cust_name":"Customer_Name"})
```

```python
df.columns
```

```python
#Gives the Descritption of numerical data in the dataset (like count, min, max, mean,median,std)
df.describe()
```

# Exploratory Data Analysis

## Gender

```python
g1 = sns.countplot(x='Gender',data = df)

for bars in g1.containers:
    g1.bar_label(bars)
```

```python
g2 = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)
```

```python
sns.barplot(x='Gender',y='Amount',data = g2)
```

By Observing the above graphs, Females are the most buyers than Males and even have more purchasing power than males

## Age

```python
#No.of. Customer by Age and Gender
a1 = sns.countplot(x='Age Group',data = df, hue = 'Gender')
a1
for bars in a1.containers:
    a1.bar_label(bars)
```

```python
#Total Amount of Sales by Age category
a2 = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Age Group',ascending =False)
sns.barplot(x='Age Group',y='Amount',data = a2)
```

## State

```python
#Top 10 States who have purchased more
s1 = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)

sns.set(rc={'figure.figsize':(20,10)})  #setting the Width and Height of the bars in order to avoid overlapping of labels

sns.barplot(y='State',x='Amount',data = s1)
```

```python
#Top 10 States who have ordered more
s1 = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)

sns.set(rc={'figure.figsize':(20,5)})  #setting the Width and height of the bars in order to avoid overlapping of labels

sns.barplot(x='State',y='Orders',data = s1)
```

* From the above Graphs, it is observed that most of the Order and Sales are happened in UP, Maharashtra and Karnataka

## Marital Status

```python
M1 = sns.countplot(data = df, x = 'Marital_Status',hue='Gender')

for bars in M1.containers:
    M1.bar_label(bars)

sns.set(rc={'figure.figsize':(5,5)})
```

```python
#Total Sales categoried by marital Status
M2 = df.groupby(['Marital_Status','Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)

sns.set(rc={'figure.figsize':(5,5)})  #setting the Width and height of the bars in order to avoid overlapping of labels

M = sns.barplot(x='Marital_Status',y='Amount',data = M2,hue= 'Gender')
```

* From the above Graphs, it is observed that most of the purchases are done by Married Women

## Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})

O1 = sns.countplot(data = df, x = 'Occupation')

for bars in M1.containers:
    M1.bar_label(bars)
```

```python
O2 = df.groupby(['Occupation'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)
```

```python
sns.barplot(x='Occupation',y='Amount',data = O2)
```

* From the above graphs, it is obvious that most of the buyers are working in IT, Healthcare and Aviation

## Product Category

```python
P1 = sns.countplot(data = df, y = 'Product_Category')

for bars in P1.containers:
    P1.bar_label(bars)
    
sns.set(rc={'figure.figsize':(15,15)})
```

```python
P2 = df.groupby(['Product_Category'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)

sns.barplot(x='Product_Category',y='Amount',data = P2)

sns.set(rc={'figure.figsize':(25,5)})
```

* From the observation, Most of the ordered category is Food

```python
P3 = df.groupby(['Product_Category'],as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)

sns.barplot(x='Product_Category',y='Orders',data = P3)

sns.set(rc={'figure.figsize':(25,5)})
```

# Conclusion

##### Most of the buyers are 26-35 Married Women from UP, Maharashtra, Karnataka working in IT, healthcare and Aviation who are more likely to buy products from food, Clothing and Electronics category
